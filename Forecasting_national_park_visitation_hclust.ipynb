```r
# Load packages

library(ggdendro)
library(ggplot2)

parks_data <- read.csv('file1.csv')
features <- c("elevation_gain", "num_reviews", "popularity")
dist_func <- function(parks_data){
  dist(parks_data[features], method="euclidean")
}

cluster_result <- hclust(dist_func(parks_data), method="complete")

ggdendrogram(cluster_result) +
    labs(title="Parks Cluster Dendrogram")

ggsave("parks_dendrogram.png")
```
