```python
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
X = df[['popularity', 'elevation_gain', 'avg_rating', 'annual_rain']]
y = df['difficulty_rating']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
param_grid = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    'kernel': ['linear', 'rbf', 'poly'],
    'class_weight': [None, 'balanced']
}

# Initialize SVM and GridSearchCV
svm_model = SVC(random_state=42)
grid_search = GridSearchCV(svm_model, param_grid, cv=3, scoring='accuracy')
grid_search.fit(X_train, y_train)
best_svm = grid_search.best_estimator_
y_pred = best_svm.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Best SVM Parameters:", grid_search.best_params_)
print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from mpl_toolkits.mplot3d import Axes3D

X, y = make_classification(n_features=3, n_redundant=0, n_informative=3, n_clusters_per_class=1, n_classes=3)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
svm = SVC(kernel='linear')
svm.fit(X_pca, y)
xx, yy = np.meshgrid(np.linspace(X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1, 50),
                     np.linspace(X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1, 50))
zz = (-svm.intercept_[0] - svm.coef_[0][0] * xx - svm.coef_[0][1] * yy) / svm.coef_[0][2]

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, zz, alpha=0.5, cmap='coolwarm')
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap="coolwarm", s=50, edgecolors='k')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D Decision Boundary by SVM')

legend = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def polynomial_transform(x):
    """Applies a polynomial transformation with r=1 and d=2 to a 2D point."""
    x1, x2 = x
    return np.array([1, np.sqrt(2)*x1, np.sqrt(2)*x2, x1**2, np.sqrt(2)*x1*x2, x2**2])

np.random.seed(42)
points = np.random.uniform(-3, 3, (10, 2))
transformed_points = np.array([polynomial_transform(x) for x in points])
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(points[:, 0], points[:, 1], color='blue')
plt.title("Original 2D Points")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(True)

# 3D plot
ax = plt.subplot(1, 2, 2, projection='3d')
ax.scatter(transformed_points[:, 3], transformed_points[:, 4], transformed_points[:, 5], color='red')
ax.set_title("Transformed Features in Higher-dimensional Space")
ax.set_xlabel("$x_1^2$")
ax.set_ylabel("$\sqrt{2}x_1x_2$")
ax.set_zlabel("$x_2^2$")
plt.tight_layout()
plt.show()
```
